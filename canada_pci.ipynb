## Predicting canada's per capita income using linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

```

```python
df = pd.read_csv('canada_per_capita_income.csv')
print(df.head())
```

```python
X = df[['year']]                     # features must be 2D
y = df['per capita income (US$)']          # target
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
year_to_predict = np.array([[2020]])
predicted_income = model.predict(year_to_predict)
print(f"Predicted Per Capita Income for 2020: {predicted_income[0]:.2f} CAD")
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(year_to_predict, predicted_income, color='green', label='Prediction (2020)')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (CAD)')
plt.legend()
plt.title('Canada Per Capita Income Prediction')
plt.show()
```
